# One Hot Encoding with Sklearn

```python
df.head()
```

**Spliting X and Y**

```python
x = df[['Present_Price','Kms_Driven','Fuel_Type','Transmission']]

y = df['Selling_Price']
```

```python
#Training and Testing the data

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=.20,random_state=1)
```

```python
x_train.head()
```

**Appling Sklearn OneHot Encoding**

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
```

```python
x_train_new = ohe.fit_transform(x_train[['Fuel_Type','Transmission']]).toarray()
x_test_new = ohe.transform(x_test[['Fuel_Type','Transmission']]).toarray()
```

```python
x_train_new
```
